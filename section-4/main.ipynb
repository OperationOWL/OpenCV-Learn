```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img(img_source):
    return cv2.cvtColor(cv2.imread(img_source).astype(np.float32)/255, cv2.COLOR_BGR2RGB)
```

```python
img = load_img(r"bricks.jpg")
```

```python
plt.imshow(img)
```

```python
gamma = 8
```

```python
pic = np.power(img, gamma)
```

```python
fig = plt.figure(figsize = (12,8))
plt.imshow(pic)
```

```python
#blurring
```

```python
font = cv2.FONT_HERSHEY_COMPLEX
img = cv2.putText(img, text = "bricks", org = (10, 600), fontFace = font,fontScale = 10, color=(255,0,0), thickness = 4)
```

```python
plt.imshow(img)
```

```python
kernel = np.ones(shape = (5,5), dtype = np.float32)/10
```

```python
kernel
```

```python
dst = cv2.filter2D(img, -1, kernel)
plt.imshow(dst)
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.blur(img, ksize=(100,100)))
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.GaussianBlur(img, (11,11), 10))
```

```python
#median blur is a technique used to remove noise from an image, and displaying only the crisp features.
```

```python
fig = plt.figure(figsize = (12,12))
plt.imshow(cv2.medianBlur(img, 7))
```
