```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread(r"dog_backpack.png")
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.flip(img, 0))
```

```python
vertices = np.array([[200, 350], [600, 350], [600, 700], [200, 700]])
vertices = vertices.reshape((-1, 1, 2))
plt.imshow(cv2.polylines(img, [vertices], isClosed=True, color=(0,0,255), thickness=20))
```
